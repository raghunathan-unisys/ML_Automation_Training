# Questions 1
1. What is the average age of the Titanic passengers?
2. What is the median age and ticket fare price of the Titanic passengers?
3. What is the average age for male versus female Titanic passengers?
4. What is the mean ticket fare price for each of the sex and cabin class
combinations?
5. What is the number of passengers in each of the cabin classes?


```python
# Importing the Titanic Dataset

import numpy as np

import seaborn as sns
print(sns.get_dataset_names())

data = sns.load_dataset("titanic")
data.head()
```

```python
# 1. Average age of Titanic Passengers
avg_age = np.mean(data['age'])

print(f"Average age of Passengers: {avg_age}")
```

```python
# 2. What is the median age and ticket fare price of the Titanic passengers?

median_age = np.median(data['age'])
print(f"Median age: {median_age}")

median_fare = np.median(data['fare'])
print(f"Median fare: {median_fare}")
```

```python
# 3. What is the average age for male versus female Titanic passengers?

age_male = data['age'][data['sex'] == 'male']
age_female = data['age'][data['sex'] == 'female']

avg_age_male = np.mean(age_male)
avg_age_female = np.mean(age_female)

print(f"Average age of Male passengers: {avg_age_male}")
print(f"Average age of Female passengers: {avg_age_female}")
```

```python
# 4. What is the mean ticket fare price for each of the sex and cabin class
# combinations?

mean_fare_sex_cabin = data.groupby(['sex', 'class'])['fare'].mean()
print(mean_fare_sex_cabin)
```

```python
# 5. What is the number of passengers in each of the cabin classes?

print(data['class'].value_counts())
```
